## Weather Forecast for a Week

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

def forecast_week():
    try:
        page = requests.get('https://forecast.weather.gov/MapClick.php?lat=35.2238&lon=-80.8411')
    except requests.exceptions.RequestException as e:
        print(f"looks like error in making call: {e}")
        return None


    try:
        soup = BeautifulSoup(page.content,'html.parser')

        seven_day_forecast = soup.find(id="seven-day-forecast")

        ## Now That you have data for 7 days of forecast clean them and capture them to build a dataframe

        period_tags = seven_day_forecast.select(".tombstone-container .period-name")
        periods = [period.get_text() for period in period_tags]
        

        short_desc_tags = seven_day_forecast.select(".tombstone-container .short-desc")
        short_desc = [short.get_text() for short in short_desc_tags]
        

        temp_tags = seven_day_forecast.select(".tombstone-container .temp")
        temps = [temp.get_text() for temp in temp_tags]

        weather_sentence_tags = seven_day_forecast.select(".tombstone-container img")
        weather_sentence = [weather['title']for weather in weather_sentence_tags]

        ## Pushing data in DataFrame
        weather = pd.DataFrame({"day": periods, "forecast":short_desc,"temperature":temps,"description": weather_sentence})
        return weather
    except Exception as e:
        print(f"Error occured while data processing: {e}")
        
    
```

```python
forecast_week()
```
